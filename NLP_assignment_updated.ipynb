Assignment 1:Perform Text Pre-processing techniques on 2 paragraphs as 1 input

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import warnings
warnings.filterwarnings('ignore')
```

Input text

```python
text = """Data Science is an interdisciplinary field that uses scientific methods, processes, algorithms, and systems to extract knowledge from data. It involves statistics, machine learning, and data analysis. Data scientists work with structured and unstructured data, and they apply various techniques to clean, process, and analyze it. However, data cleaning: removing noise, handling missing values, and correcting inconsistencies, is one of the most crucial steps! Without proper preprocessing, models may give inaccurate results.

Natural Language Processing (NLP) is a branch of artificial intelligence; it focuses on enabling machines to understand, interpret, and generate human language. NLP combines computational linguistics with machine learning techniques. Can machines truly understand human language? Yes, they can! Applications such as chatbots, sentiment analysis, and language translation are powered by NLP systems, which continue to evolve rapidly."""
```

```python
text
```

Removing Punctuation 

```python
text_no_punct = text.translate(str.maketrans('', '', string.punctuation))
print(text_no_punct)
```

Word tokenization and Converting text to lowercase

```python
text = word_tokenize(text_no_punct.lower())
print(text)
```

Remove Stop Words from text 

```python
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in text if word not in stop_words]
print(filtered_words)
```

Apply Stemming on my text

```python
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]
print(stemmed_words)
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
print(lemmatized_words)
```

Assignment 2: Implement Feature engineering steps ( Bag of Words, TF-IDF) on same input.

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

Bag of words(BOW)

```python
clean_text = " ".join(lemmatized_words)
corpus = [clean_text]
cv = CountVectorizer()
bow = cv.fit_transform(corpus)
print("Feature Names:", cv.get_feature_names_out())
print("BoW Matrix:", bow.toarray())
```

TF-IDF

```python
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(corpus)
print("Feature Names:", tfidf.get_feature_names_out())
print("TF-IDF Matrix:", tfidf_matrix.toarray())
```

Input data

```python
original_text2 = """Data science involves statistics and analysis to extract meaningful insights and findings from data. 
Correlation and relationship between variables help in understanding patterns, while regression and predictive analysis are used for trend modeling and forecasting. 
Classification or categorization techniques are applied for label prediction, and clustering or segmentation methods are used for grouping similar data points.

Businesses rely on dashboards and data visualization panels to monitor KPIs, also known as metrics or business indicators, to evaluate performance. 
Customer churn, also called customer attrition or drop-off rate, is an important problem that companies try to solve using machine learning models. 
Recommendation systems, also known as suggestion engines or personalization systems, help improve user experience by providing relevant content.

In programming, scripts or code are written using libraries, also known as packages or modules, and frameworks, which are development platforms or toolkits. 
During development, bugs or errors, also called issues, are common and need to be fixed. 
Overall, data science combines programming, statistical analysis, and machine learning to generate actionable insights and support decision-making processes."""
```

Converting text into sentences

```python
from nltk.tokenize import sent_tokenize
sentences = sent_tokenize(original_text2)
print(sentences)
```

Remove Punctuations

```python
import re
clean_sentences = []
for sentence in sentences:
    clean = re.sub(r'[^a-zA-Z ]', '', sentence).lower()
    clean_sentences.append(clean)
print(clean_sentences)
```

Converting Sentences to Vectors

```python
import numpy as np
def sent_vec(sentence):
  words = sentence.lower().split()
  vector = []
  for word in words:
    if word in model.wv:
      vector.append(model.wv[word])
  return np.mean(vector, axis=0)
```

```python
sentence_vectors = [sent_vec(sentence) for sentence in clean_sentences]
sentence_vectors = np.array(sentence_vectors)
print(sentence_vectors)
```

Cosine similarity between sentences

```python
from sklearn.metrics.pairwise import cosine_similarity
s1=sent_vec("data science involves statistics and analysis to extract meaningful insights and findings from data")
s2=sent_vec("correlation and relationship between variables help in understanding patterns while regression and predictive analysis are used for trend modeling and forecasting")
s3=sent_vec("classification or categorization techniques are applied for label prediction and clustering or segmentation methods are used for grouping similar data points")
similarity_1 = cosine_similarity([s1],[s2])
similarity_2 = cosine_similarity([s1],[s3])
similarity_3 = cosine_similarity([s2],[s3])
print("Sentence 1 and 2 Similarity:")
print(similarity_1[0][0])
print("Sentence 1 and 3 Similarity:")
print(similarity_2[0][0])
print("Sentence 2 and 3 Similarity:")
print(similarity_3[0][0])
```
